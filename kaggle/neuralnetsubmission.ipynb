```python
from torch.utils.data import Dataset
import numpy as np
```

```python
training_data = torch.tensor(training_data.astype(float))
```

```python
test_data = torch.tensor(test_data.astype(float))
```

```python
class my_data(Dataset):
  def __init__(self,training_data,transform = None,target_transform = None):
    self.data = training_data
    self.transform = transform;
    self.target_transform = target_transform;

  def __len__(self):
    return self.data.shape[0];

  def __getitem__(self,idx):
    train = self.data[idx,0:75]
    label = self.data[idx,75]
    if self.transform:
      train = self.transform(train)

    if self.target_transform:
      label = self.target_transform(label)
      
    return train, label
```

```python
trainset = my_data(training_data,transform=None,target_transform = None)
#testset = my_data(test_data,test_label,transform = transform)
```

```python
trainloader = torch.utils.data.DataLoader(trainset,batch_size = 64,shuffle = True)
#testloader = torch.utils.data.DataLoader(trainset,batch_size = 64,shuffle = True)
```

```python
from torch import nn
model = nn.Sequential(nn.Linear(75,256),
                      nn.ReLU(),
                      nn.Dropout(p=0.2),
                      nn.Linear(256,256),
                      nn.ReLU(),
                      nn.Linear(256,128),
                      nn.ReLU(),
                      nn.Dropout(p=0.2),
                      nn.Linear(128,64),
                      nn.ReLU(),
                      nn.Dropout(p=0.2),
                      nn.Linear(64,9),
                      nn.LogSoftmax(dim = 1));
print(model)
```

```python
feature, label = next(iter(trainloader))
output = model(feature.float())
out = torch.argmax(output, dim = 1)
print(output)
```

```python
from torch import optim
optimizer = optim.SGD(model.parameters(),lr = 0.002)
```

```python
print(label.shape)
```

```python
with torch.no_grad():
    input = test_data
    output = model(input[:,0:75].float())
    _, predicted = torch.max(output,dim = 1)
    equals = predicted==test_data[:,75]
    #print(equals.shape)
    correct += torch.sum(torch.tensor(equals)).item()
    #loss = criterion(output,test_data[:,74])
    #test_running_loss += loss
    #test_loss.append(test_running_loss)
    #print(f"test_loss: {test_running_loss}")
    print(f"Accuracy: {correct/40000}")
```

```python

epochs = 200
criterion = nn.NLLLoss()
for e in range(epochs):
  running_loss = 0
  for features, labels in iter(trainloader):
    optimizer.zero_grad()
    output = model(features.float())
    #predicted = torch.argmax(output,dim = 1,keepdim = False)
    labels = labels.type(torch.LongTensor)
    loss = criterion(output,labels)
    running_loss+= loss
    loss.backward()
    optimizer.step()
  #training_loss.append(running_loss)
  print(f"Loss = {running_loss}")
  correct = 0
  test_running_loss = 0
  with torch.no_grad():
    input = test_data
    output = model(input[:,0:75].float())
    _, predicted = torch.max(output,dim = 1)
    equals = predicted==test_data[:,75]
    #print(equals.shape)
    correct += torch.sum(torch.tensor(equals)).item()
    #loss = criterion(output,test_data[:,74])
    #test_running_loss += loss
    #test_loss.append(test_running_loss)
    #print(f"test_loss: {test_running_loss}")
    print(f"Accuracy: {correct/40000}")
```
